```python
def random_date_generator(start_date, range_in_days, days_to_add):
    random_date = np.datetime64(start_date) + days_to_add
    return random_date
```

```python
#'start_date' is anything between 2018 and december 31 2021
#'end_date' is anything between start_date and april 14th 2023 (make this a datetime object)

start_dates = []
end_dates = []
for i in range(200):
  days_to_add = np.random.choice(np.arange(0, 1929))
  start_dates.append(random_date_generator("2018-01-01", 1929, days_to_add))
  days_to_add_end = np.random.choice(np.arange(days_to_add, 1929))
  end_dates.append(random_date_generator("2018-01-01", 1929, days_to_add_end))
```
